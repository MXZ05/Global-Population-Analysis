**Bar Chart for the Top 20 Populations in 2020**

```python
population_2020 = df[['Country Name','2020']].dropna()
population_2020 = population_2020.sort_values(by='2020',ascending=False).head(20)

# Convert population to billions
population_2020['2020'] = population_2020['2020'] / 1e9

population_2020
```

```python
import matplotlib.pyplot as plt
import matplotlib.style as style

# Apply dark background style
style.use('dark_background')
```

```python
plt.figure(figsize=(12,8))
plt.bar(population_2020['Country Name'],population_2020['2020'],color='lightcoral')
plt.title('Top 20 Populations in 2020',fontsize=16)
plt.xlabel('Country/Region/Class',fontsize=12)
plt.ylabel('Population in Billions',fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.tight_layout()

# Show the plot
plt.show()
```

**LINE CHART OF POPULATION GROWTH**

```python
# Select data for multiple countries and years
countries = ['China', 'India', 'Pakistan','United States']
population_trends = df[['Country Name'] + [str(year) for year in range(1960, 2024)]]

# Filter data for selected countries
population_trends = population_trends[population_trends['Country Name'].isin(countries)]

# Plot the line chart for each country
plt.figure(figsize=(10,6))

for country in countries:
    plt.plot(population_trends.columns[1:], population_trends[population_trends['Country Name'] == country].iloc[0,1:],label=country)

# Customize the chart
plt.title('Population Growth Trends (1960-2020)',fontsize=16)
plt.xlabel('Year',fontsize=14)
plt.ylabel('Population in Billions',fontsize=14)
plt.legend(title='Countries')
plt.xticks(ticks=range(0, len(population_trends.columns[1:]), 5), labels=population_trends.columns[1:][::5])
plt.tight_layout()
plt.tight_layout()
plt.show()

```

**Histogram for Population Distribution**

```python
plt.figure(figsize=(10,6))
plt.hist(population_2020['2020'], bins=15,color='lightcoral',edgecolor='black')
plt.title('Population Distribution Across Countries in 2020',fontsize=16)
plt.xlabel('Population',fontsize=12)
plt.ylabel('Number of Countries',fontsize=12)
plt.tight_layout()

plt.show()
```
